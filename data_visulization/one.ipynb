```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### matplotlib basic

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("x and y")
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### subplots in matplotlib


```python
plt.subplot(2,2,1) # hum jo plot kiyea hai ushme fig v draw kar skte hai bus hum code niche likhate 
plt.plot(x,y)# hai taki ushi plot me figure ban sake 
plt.subplot(2,2,4)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,y**2)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

## object oriented way

```python
fig=plt.figure(figsize=(10,4),dpi=500)  # dpi use karne se number clear dikhane lagta hai
axis1=fig.add_axes([0.1,0.1,0.4,.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,.3])
axis2.plot(x,y)
```
