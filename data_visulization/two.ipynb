### let's discuss some plots that allow us to distribution of a data set.There plots are/
displot/histolot\
.jointplot\
.pairplot\
.rugplot

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex') # kind= reg v kar sakte hai 
```

# pairplot


```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow') # color se thoda graph accah dikhata hai 
```

### Rug plot

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df) # ishme pura data ka rug plot ban gya hai 
```

### Categorical Data plots 
.box plot\
.violin plot\
.stripp plot\
.swarm plot\
.bar plot\
count plot\

```python
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

### Bar plot

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

### Boxplot

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

### violion plot

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

### stripplot and swarmplot

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df) # do plots ko merge kar denge ek hi figure me 
sns.swarmplot(x='tip',y='day',data=df)
```

### Heat map

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

### Cluster map


```python
sns.clustermap(tipscorr.corr())
```

### Piviot table  heatmap

```python
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
